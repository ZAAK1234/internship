```python
urls=[]
ser_list=['fruits','cars','Machine Learning','Guitar','cakes']
for i in ser_list:
    dr.get('https://images.google.com/')
    ser_bar=dr.find_element(By.XPATH,'//textarea[@class="gLFyf"]')
    ser_bar.send_keys(i)
    try:
        ser_btn=dr.find_element(By.XPATH,"//button[@class='Tg7LZd']")
        ser_btn.click()
    except ElementClickInterceptedException:
        ser_btn1=dr.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span/svg')
        ser_btn1.click()
    time.sleep(5)
    img=dr.find_elements(By.XPATH,'//img[@class="rg_i Q4LuWd"]')
    for j in img[0:10]:
        im=j.get_attribute('src')
    
        urls.append(im)
```
