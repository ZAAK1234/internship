### Question 4
Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com 
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,  
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
details is missing then replace it by “- “. Save your results in a dataframe and CSV. 

```python
from bs4 import BeautifulSoup
import requests
import selenium
import time
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
```

```python
dr=webdriver.Chrome()
dr.get('http://www.flipkart.com/')
```

```python
dr.get('http://www.flipkart.com/')
try:
    log_rmvl=dr.find_element(By.XPATH,'//span[@class="_30XB9F"]').click()
except NoSuchElementException as e:
    print('Exception Raised',e)
    

```

```python
ser_bar=dr.find_element(By.XPATH,'//input[@class="Pke_EE"]')
ser_bar.send_keys('moto g32')

ser_btn=dr.find_element(By.XPATH,'//button[@class="_2iLD__"]')
ser_btn.click()
```

```python
urls=[]
Brand=[]
Model=[]
colour=[]
Ram=[]
Storage=[]
P_camera=[]
S_camera=[]
Dspls=[]
Btrcp=[]
Price=[]


url=dr.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in url:
    #dr.get('http://www.flipkart.com/')
    op=i.get_attribute('href')
    urls.append(op)
```

```python
len(urls)
```

```python
for i in urls:
    dr.get(i)
    time.sleep(5)
    
    rd_mr=dr.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]').click()
    try:
        brand=dr.find_element(By.XPATH,"//span[@class='B_NuCI']").text
        Brand.append(brand)
        time.sleep(3)
    except:
        Brand.append('-')
    try:
        smrt_ph=dr.find_element(By.XPATH,'//div[@class="_3k-BhJ"]/table/tbody/tr[3]/td[2]/ul/li').text
        Model.append(smrt_ph)
    except:
        smrt_ph.append('-')
    try:
        clr=dr.find_element(By.XPATH,'//div[@class="_3k-BhJ"]/table/tbody/tr[4]/td[2]/ul/li').text
        colour.append(clr)
    except:
        colour.append('-')
    try:
        rm=dr.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[4]/table/tbody/tr[2]/td[2]/ul/li').text
        Ram.append(rm)
    except:
        Ram.append('-')
    try:
        strg=dr.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[4]/table/tbody/tr[1]/td[2]/ul/li').text
        Storage.append(strg)
    except:
        Storage.append('-')
    try:
        p_camera=dr.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[5]/table/tbody/tr[2]/td[2]/ul/li').text
        P_camera.append(p_camera)
    except:
        P_camera.append(p_camera)
    try:
        s_camera=find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[5]/table/tbody/tr[5]/td[2]/ul/li').text
        S_camera.append(s_camera)
    except:
        S_camera.append('-')
    try:
        dspl_s=dr.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[2]/table/tbody/tr[1]/td[2]/ul/li').text
        Dspls.append(dspl_s)
    except:
        Dspls.append('-')
    try:
        btrcp=dr.find_element(By.XPATH,'//div[@class="_1UhVsV"]/div[10]/table/tbody/tr[1]/td[2]/ul/li').text
        Btrcp.append(btrcp)
    except:
        Btrcp.append('-')
    try:
        prc=dr.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]').text
        Price.append(prc)
    except:
        Price.append('-')
    
```

```python
smartphone=pd.DataFrame({})
smartphone['Brand']=Brand
smartphone['Model']=Model
smartphone['Color']=colour
smartphone['Ram']=Ram
smartphone['Storage']=Storage
smartphone['Primary camera']=P_camera
smartphone['Secondary Camera']=S_camera
smartphone['Display Size']=Dspls
smartphone['Battery Capicity']=Btrcp
smartphone['Price']=Price
```
