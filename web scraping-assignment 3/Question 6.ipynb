### Question 6
Write a program to scrap all the available details of best gaming laptops from digit.in. 

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
 dr=webdriver.Chrome()
dr.get('https://digit.in/')
```

```python
urls=[]
Brand_urls=[]
Brand=[]
Op_stm=[]
Dspls=[]
Rsltn=[]
Prcsr=[]
Price=[]
url=dr.find_elements(By.XPATH,'//div[@class="header_icons_menu search-form-inheader main-nav mob-logo-enabled dark_style"]/div/nav/ul/li[2]/ul/li[3]/ul/li/a')
url=dr.find_elements(By.XPATH,'//div[@class="header_icons_menu search-form-inheader main-nav mob-logo-enabled dark_style"]/div/nav/ul/li[2]/ul/li[3]/ul/li/a')
for i in url:
    l=i.get_attribute('href')
    urls.append(l)
for j in urls:
    dr.get(j)
    time.sleep(10)
    brnd_urls=dr.find_elements(By.XPATH,'//h3[@class="font130 mt0 mb10 mobilesblockdisplay "]/a')
    for k in brnd_urls:
        Br_url=k.get_attribute('href')
        Brand_urls.append(Br_url)
    rsltn_=dr.find_elements(By.XPATH,'//div[@class="woo_code_zone_loop clearbox"]/div[3]/div[1]/span[2]')
    for d in rsltn_:
        Rsltn.append(d.text)
        time.sleep(3)
        time.sleep(10)
for i in urls:
    dr.get(i)
    time.sleep(5)
    brnd_urls=dr.find_elements(By.XPATH,'//h3[@class="font130 mt0 mb10 mobilesblockdisplay "]/a')
for j in Brand_urls:
    dr.get(j)
    time.sleep(5)
    brnd=dr.find_elements(By.XPATH,'//h1[@class="floatleft tabletblockdisplay pr20 "]')
    for a in brnd:        
        Brand.append(a.text)
        time.sleep(5)
    opstm=dr.find_elements(By.XPATH,'//div[@class="specs_details "]/ul/li[1]/div/p[2]')
    for b in opstm:    
        Op_stm.append(b.text)
        time.sleep(3)
    dspl=dr.find_elements(By.XPATH,'//div[@class="specs_details "]/ul/li[4]/div/p[2]')
    for c in dspl:
        Dspls.append(c.text)
        time.sleep(3)
    
    prcsr_=dr.find_elements(By.XPATH,'//div[@class="specs_details "]/ul/li[2]/div/p[2]')
    for e in prcsr_:
        Prcsr.append(e.text)
        time.sleep(3) 
    try:
        price_=dr.find_element(By.XPATH,'//div[@class="top_share_small top_share notextshare"]/div/div/div[2]/div[3]/div/div[1]') 
        Price.append(price_.text)
        time.sleep(3)
    except NoSuchElementException:
        Price.append('-')
```

```python
Laptop=pd.DataFrame({})
Laptop['Brand']=Brand
Laptop['Operating system']=Op_stm
Laptop['Display Size']=Dspls
Laptop['Resolution']=Rsltn
Laptop['Processor']=Prcsr
Laptop['Price']=Price
Laptop
```
