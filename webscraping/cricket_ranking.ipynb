```python
from bs4 import BeautifulSoup
import requests
```

```python
url = requests.get('https://www.icc-cricket.com/rankings/team-rankings/mens/odi')
url
```

```python
bs=BeautifulSoup(url.content)
bs
```

```python
teams=[]
for i in bs.find_all('h3',class_="si-team-name"):
    teams.append(i.text)
teams
```
