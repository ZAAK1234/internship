```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dr=webdriver.Chrome()
dr.get('https://www.azquotes.com/')
```

```python
tp_qzbtn=dr.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
tp_qzbtn.click()
```

```python
qz_title=[]
qz_auther=[]
qz_type=[]
```

```python
for i in range(0,10):
    title=dr.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')
    for j in title[0:100]:
        qtitle=j.text
        qz_title.append(qtitle)
    nxtbtn=dr.find_element(By.CLASS_NAME,'next')
    nxtbtn.click()
    time.sleep(3)
print(len(qz_title))
print(qz_title)
```

```python
for i in range(0,10):
    auther=dr.find_elements(By.XPATH,'//div[@class="author"]/a[1]')
    for j in auther[0:100]:
        qauther=j.text
        qz_auther.append(qauther)
    nxtbtn=dr.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[13]')
    nxtbtn.click()
print(len(qz_auther))
print(qz_auther)
```

```python
for i in range(0,10):
    qztype=dr.find_elements(By.XPATH,'//div[@class="tags"]')
    for j in qztype[0:100]:
        qtype=j.text
        qz_type.append(qtype)
    nxtbtn=dr.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[13]')
    nxtbtn.click()
print(len(qz_type))
print(qz_type)
```

```python
df=pd.DataFrame({'Quote':qz_title,'Auther':qz_auther,'Type of Quote':qz_type})
df
```
